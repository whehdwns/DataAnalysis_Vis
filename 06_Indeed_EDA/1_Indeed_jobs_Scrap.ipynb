```python
from selenium import webdriver
from bs4 import BeautifulSoup
```

```python
job_title ='data+analyst'
```

```python
url_list =[]
for i in range(0, 51):
    url_list.append("https://www.indeed.com/jobs?q="+job_title+"&start="+str(i*10))
```

```python
job_title_list =[]
job_company_list =[]
job_link_list=[]

for i in url_list:
    driver = webdriver.Firefox()
    driver.get(i)
    page = driver.page_source
    soup = BeautifulSoup(page, 'html.parser')
    job_description = soup.find_all('ul', class_='jobsearch-ResultsList css-0')
    joblisted = job_description[0].find_all('td', class_='resultContent')

    for j in range(len(joblisted)):
        job_title = joblisted[j].find_all('span')[0].text
        job_company = joblisted[j].find_all('span')[1].text
        job_link = 'https://www.indeed.com' + joblisted[j].find('a')['href']

        job_title_list.append(job_title)
        job_company_list.append(job_company)
        job_link_list.append(job_link)
    driver.quit()
    print(i)
```
